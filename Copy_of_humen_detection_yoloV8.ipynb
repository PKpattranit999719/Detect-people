```python
# à¹‚à¸«à¸¥à¸”à¹„à¸Ÿà¸¥à¹Œà¸œà¸¥à¸¥à¸±à¸à¸˜à¹Œà¸ˆà¸²à¸à¸à¸²à¸£à¸à¸¶à¸
metrics = pd.read_csv('/content/runs/detect/train2/results.csv')

# à¸•à¸£à¸§à¸ˆà¸ªà¸­à¸šà¸‚à¹‰à¸­à¸¡à¸¹à¸¥
print(metrics.head())
```

```python
plt.figure(figsize=(10, 6))

# à¸à¸£à¸²à¸Ÿ Train Loss
plt.plot(metrics['epoch'], metrics['train/box_loss'], label='Train Box Loss', linestyle='-', marker='o')
plt.plot(metrics['epoch'], metrics['train/cls_loss'], label='Train Class Loss', linestyle='-', marker='o')

# à¸à¸£à¸²à¸Ÿ Validation Loss
plt.plot(metrics['epoch'], metrics['val/box_loss'], label='Validation Box Loss', linestyle='--', marker='x')
plt.plot(metrics['epoch'], metrics['val/cls_loss'], label='Validation Class Loss', linestyle='--', marker='x')

plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Train and Validation Loss')
plt.legend()
plt.grid(True)
plt.show()
```
